```python
data['timestamp'] = pd.to_datetime(data['timestamp'],dayfirst=True)
```

```python
data['Month'] =  data['timestamp'].dt.month
data['Month_num'] =  data['Month'].apply(lambda x : int(x))
data['Month_name'] =  data['Month_num'].apply(lambda x: calendar.month_abbr[x])
data['Year'] =  data['timestamp'].dt.year
data['Year_num'] = data['Year'].apply(lambda x : int(x))
data['Day'] = data['timestamp'].dt.day
data['Day_of_Week'] = data['timestamp'].dt.day_name()
data.drop(columns='Year', inplace=True)
data.drop(columns='Month', inplace=True)
```

```python
# Extract the hour component from the timestamp
data['Hour'] = data['timestamp'].dt.hour

# Create a function to categorize the time of day
def categorize_time(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    elif 18 <= hour < 24:
        return "Evening"
    else:
        return "Night"

# Apply the function to create a new 'Time_of_Day' column
data['Time_of_Day'] = data['Hour'].apply(categorize_time)
```
